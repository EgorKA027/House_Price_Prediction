# ID

```python
df.drop(['ID', 'Тип', 'Телефоны', 'Описание'], axis = 1, inplace = True)
```

```python
df.head(3)
```

```python
import seaborn as sns
data = pd.DataFrame()
data.index = df.columns
data['dtypes'] = df.dtypes
data['NaN'] = df.isna().sum()
data['count'] = df.count()
data['NaN%'] = round((data['NaN'] / df.shape[0]) * 100, 2)

data.style.background_gradient(cmap='Blues')
```

```python
columns = data[data['NaN%'] < 41].index
df = df[columns]
df.shape
```

### Метро

```python
df['Метро'] = df['Метро'].apply(lambda x: str(x).replace(')', '')).copy()
df['Метро'] = df['Метро'].apply(lambda x: str(x).split(' (')).copy()
```

### Площадь

```python
df['Площадь, м2'] = df['Площадь, м2'].apply(lambda x: x.split('/')).copy()
```

### Дом

```python
df['Дом'] = df['Дом'].apply(lambda x: x.split(',')).copy()
```

### Цена

```python
df['Цена'] = df['Цена'].apply(lambda x: x.replace('.','')).copy()
df['Цена'] = df['Цена'].apply(lambda x: x.split(' руб')).copy()
```
